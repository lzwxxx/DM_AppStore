## **Data Cleaning**

```python
df.drop_duplicates(['App', 'Size', 'Installs', 'Reviews', 'Rating', 'Price', 'Android Ver'], inplace=True)          
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.describe()
```

```python
def value_to_int(x):
    if isinstance(x, str) and 'M' in x:
        if len(x) > 1:
            return int(float(x.replace('M', '')) * 1000000)
        return 1000000
    elif isinstance(x, float):
        return int(x)
    else:
        return x
    
df['Reviews'] = df['Reviews'].apply(value_to_int)
```

```python
df["Reviews"] = pd.to_numeric(df["Reviews"])
```

```python
# REMOVE "+" 
df.Size = df["Size"].str.replace("+", "")

# REMOVE "," 
df.Size = df["Size"].str.replace(",", "")
```

```python
#Converting values with "k" and "M" into numbers 
def value_to_int(x):
    if 'k' in x:
        if len(x) > 1:
            return int(float(x.replace('k', '')) * 1000)
        return 1000
    if 'M' in x:
        if len(x) > 1:
            return int(float(x.replace('M', '')) * 1000000)
        return 1000000
    if type(x) == float or type(x) == int or type(x) == str:
        return x

df['Size'] = df['Size'].apply(value_to_int)
```

```python
df['Size'].replace('Varies with device', np.nan , inplace = True)
```

```python
df['Size'].fillna(df.groupby('Category')['Size'].transform('mean'), inplace = True)
```

```python
df["Size"] = df["Size"].astype(float)
```

```python
# CONVERT "INSTALLS" COL INTO FLOAT/INT (INITIALLY: OBJECT)

# REMOVE "+" 
df.Installs = df["Installs"].str.replace("+", "")

# REMOVE "," 
df.Installs = df["Installs"].str.replace(",", "")

#There is a "free" value in the Installs column
df.Installs = pd.to_numeric(df["Installs"])
```

```python
# CONVERT PRICE TO FLOAT

# REMOVE "$"
df.Price = df["Price"].str.replace("$", "")

df.Price = df["Price"].astype(float)
```

```python
df['Genre 1']= df["Genres"].apply(lambda x: x.split(";")[0] if ";" in x else x)
df['Genre 2']= df["Genres"].apply(lambda x: x.split(";")[1] if ";" in x else "-")

# df.drop('Genres', axis = 1, inplace = True)
df.head()
```

```python
import datetime

# Converting into YYYYMM format, and drop exact dates
df['Last Updated'] = df['Last Updated'].apply(lambda x : datetime.datetime.strptime(x, '%B %d, %Y').strftime('%Y%m')).astype(int)
```

```python
import re
# Clean non numerical values & unicode characters 
replaces = [u'\u00AE', u'\u2013', u'\u00C3', u'\u00E3', u'\u00B3', '[', ']', "'"]
for i in replaces:
	df['Current Ver'] = df['Current Ver'].astype(str).apply(lambda x : x.replace(i, ''))

regex = [r'[-+|/:/;(_)@]', r'\s+', r'[A-Za-z]+']
for j in regex:
	df['Current Ver'] = df['Current Ver'].astype(str).apply(lambda x : re.sub(j, '0', x))

df['Current Ver'] = df['Current Ver'].astype(str).apply(lambda x : x.replace('.', ',',1).replace('.', '').replace(',', '.',1)).astype(float)
# filling null values with median
df['Current Ver'] = df['Current Ver'].fillna(df['Current Ver'].median())
```

```python
import re
# Clean non numerical values & unicode characters 
replaces = [u'\u00AE', u'\u2013', u'\u00C3', u'\u00E3', u'\u00B3', '[', ']', "'"]
for i in replaces:
	df['Android Ver'] = df['Android Ver'].astype(str).apply(lambda x : x.replace(i, ''))

regex = [r'[-+|/:/;(_)@]', r'\s+', r'[A-Za-z]+']
for j in regex:
	df['Android Ver'] = df['Android Ver'].astype(str).apply(lambda x : re.sub(j, '0', x))

df['Android Ver'] = df['Android Ver'].astype(str).apply(lambda x : x.replace('.', ',',1).replace('.', '').replace(',', '.',1)).astype(float)
# filling null values with median
df['Android Ver'] = df['Android Ver'].fillna(df['Android Ver'].median())
```
